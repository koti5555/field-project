```python
from transformers import pipeline
```

```python
n = pipeline("sentiment-analysis")
```

```python
text = input("Enter the text: ")
print(n(text))
```

```python
gen = pipeline("text-generation")
```

```python
text = input("Enter the text: ")
print(gen(text,max_length=200))
```

```python
text = input("Enter the text:")
result = n(text)
result
```

```python
qa = pipeline("question-answering")
```

```python
c = '''Bity bought some butter,but the butter was bitter,so Bity bought some better butter to replace it.'''
```

```python
q = "what does Bitty bought?"
```

```python
result = qa({'question':q,'context': c})
print(result)
```

```python
s = pipeline("summarization")
```

```python
t = '''Bity bought some butter,but the butter was bitter,so Bity bought some better butter to replace it.'''
```

```python
result = s(t)
result[0].get('summary_text')
```
